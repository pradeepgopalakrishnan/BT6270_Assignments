```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
def run(Iext, v0, w0, a, b, r, niter=1000, dt=0.1):
    vhist=np.zeros(niter)
    whist=np.zeros(niter)
    fhist=np.zeros(niter)
    vhist[0]=v0
    whist[0]=w0
    fhist[0]=calc_f(vhist[0],a)
    t=np.arange(0,niter*dt, dt)
    for i in range(1,niter):
        w_=whist[i-1]
        v_=vhist[i-1]
        f_=fhist[i-1]
        v=v_+(f_-w_+Iext)*dt
        w=w_+(b*v_-r*w_)*dt
        f=calc_f(v_,a)
        vhist[i]=v
        whist[i]=w
        fhist[i]=f
    return vhist, whist, fhist, t
```

```python
def phaseplot(v1,v2,w1,w2,size):
    v=np.arange(v1,v2,size)
    w=np.arange(w1,w2,size)
    yy,xx=np.meshgrid(v,w)
    wd=[]
    vd=[]
    for x,y in itertools.product(v,w):
            wd.append(w_(x, y, b, r))
            vd.append(v_(x, y, Iext, a))
    wd=np.asarray(wd)
    vd=np.asarray(vd)
    return(xx,yy,vd,wd)
    #plt.quiver(xx,yy,vd,wd,scale_units='xy')
    
def phaseplot_(v1,v2,w1,w2,size):
    v=np.arange(v1,v2,size)    
    w=np.arange(w1,w2,size)
    yy,xx=np.meshgrid(v,w)
    wd=[]
    vd=[]
    for x in v:
        for y in w:
            wd.append(w_(x, y, b, r))
            vd.append(v_(x, y, Iext, a))            
    wd=np.asarray(wd)
    vd=np.asarray(vd)
    return(xx,yy,vd,wd)

#plot(-1,1.01,-1,1.01,0.1)
```

```python
def calc_f(v,a):
    return v*(a-v)*(v-1)

def w_(v, w, b, r):
    return b*v-r*w

def v_(v, w, Iext, a):
    return calc_f(v,a)-w+Iext
```
